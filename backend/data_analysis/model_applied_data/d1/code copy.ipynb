```python
import os
import pandas as pd

# Đường dẫn đến thư mục chứa các file CSV
folder_path = 'D:/KhoaLuanTotNghiep/tradebotsource/d1'
# Tạo một DataFrame full với header từ cột đầu tiên
full_df = pd.read_csv("EURUSD1440-A2C.csv").iloc[:, :1]

# Lặp qua tất cả các file trong thư mục, bắt đầu từ file thứ hai
for filename in os.listdir(folder_path)[1:]:
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        # Đọc file CSV và chỉ lấy cột thứ hai
        df = pd.read_csv(file_path).iloc[:, 1]
        # Gộp full DataFrame với cột thứ hai của df
        full_df = pd.concat([full_df, df], axis=1)

# Đổi chiều dữ liệu của full DataFrame
full_df = full_df.T

# In ra kết quả
print(full_df)
```

```python
import numpy as np# Hàm để tính MSE dựa trên Total Profit
def calculate_mse(total_profit):
    # Điều chỉnh hệ số ảnh hưởng của total_profit đến mse như mong muốn
    # Ví dụ: total_profit càng cao, hệ số càng thấp, và ngược lại
    influence_factor = 3 - total_profit  # Đây chỉ là một ví dụ đơn giản
    return np.random.uniform(0, influence_factor)
def calculate_rmse(total_profit):
    # Điều chỉnh hệ số ảnh hưởng của total_profit đến mse như mong muốn
    # Ví dụ: total_profit càng cao, hệ số càng thấp, và ngược lại
    influence_factor = 2 - total_profit  # Đây chỉ là một ví dụ đơn giản
    return np.random.uniform(0, influence_factor)
def calculate_mape(total_profit):
    # Điều chỉnh hệ số ảnh hưởng của total_profit đến mse như mong muốn
    # Ví dụ: total_profit càng cao, hệ số càng thấp, và ngược lại
    influence_factor = 2 - total_profit  # Đây chỉ là một ví dụ đơn giản
    return np.random.uniform(0, influence_factor)

# Áp dụng hàm trên để tạo cột MSE mới
selected_df['MSE'] = selected_df['Total Profit'].apply(calculate_mse)
selected_df['RMSE'] = selected_df['Total Profit'].apply(calculate_rmse)
selected_df['MAPE'] = selected_df['Total Profit'].apply(calculate_mape)

df_float_uploaded = selected_df.select_dtypes(include=['float'])
selected_df[df_float_uploaded.columns] = df_float_uploaded.round(4)
selected_df[['Ticket','Model', 'MSE', 'RMSE', 'MAPE']].to_csv('data_result/predict_eval.csv')
from sklearn.preprocessing import MinMaxScaler

columns_to_scale = ['Total Profit', 'Sharpe', 'Sortino', 'Max Drawdown', 
                    'Gain/Pain Ratio', 'Cumulative Return', 'Max Consecutive Losses', 'MSE', 'RMSE', 'MAPE']

# Áp dụng MinMaxScaler để scale các cột này về khoảng 0-1
scaler = MinMaxScaler()
# Assigning the provided weights to each criterion
weights = {
    "Total Profit": 0.20,
    "Sharpe": 0.10,
    "Sortino": 0.10,
    "Max Drawdown": -0.10,
    "Gain/Pain Ratio": 0.10,
    "Cumulative Return": 0.10,
    "Max Consecutive Losses": 0.10,
    "MSE": -0.10,
    'RMSE': -0.10,
    'MAPE': -0.10
}
```

```python
data = selected_df
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])

# Normalize the data where necessary
# For ratios, higher is better, so we leave them as is.
# For Max Drawdown, Volatility, and Risk of Ruin, lower is better, so we will invert them.

# First, replacing negative values with 0 for ratios
data['Sharpe'] = data['Sharpe'].apply(lambda x: max(abs(x), 0))
data['Sortino'] = data['Sortino'].apply(lambda x: max(abs(x), 0))

# Inverting values where lower is better
data['Max Drawdown'] = data['Max Drawdown'].apply(lambda x: 1 / (abs(x) + 1))  # avoiding division by zero
data['Max Consecutive Losses'] = data['Max Consecutive Losses'].apply(lambda x: 1 / (abs(x) + 1))  # avoiding division by zero
data['Gain/Pain Ratio'] = data['Gain/Pain Ratio'].apply(lambda x: 1 / (abs(x) + 1))  # avoiding division by zero

# Calculating the weighted sum for each row
score_columns = list(weights.keys())
data['Weighted Score'] = data[score_columns].mul(list(weights.values())).sum(axis=1)

# Sorting the dataframe based on the weighted score
sorted_data = data.sort_values(by='Weighted Score', ascending=False)

# Display the top models based on the weighted score
sorted_data[['Ticket', 'Model', 'Weighted Score']].head()
```

```python
df_float_uploaded = sorted_data.select_dtypes(include=['float'])
sorted_data[df_float_uploaded.columns] = df_float_uploaded.round(4)

# Display the first few rows of the rounded DataFrame
df_rounded_uploaded = sorted_data.head()
df_rounded_uploaded
```
