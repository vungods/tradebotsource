```python
import pandas as pd
data=pd.read_csv('predict_eval_m5.csv')
# Splitting the 'Ticket' column into two parts
data['Ticket Code'] = data['Ticket'].str.extract(r'([A-Za-z]+)')
data['Period Time'] = data['Ticket'].str.extract(r'(\d+)')

# Convert 'Period Time' to integer
data['Period Time'] = data['Period Time'].astype(int)

# Display the updated dataframe
# Removing the numeric part (1440) from the 'Ticket' column
data['Ticket'] = data['Ticket Code']
data['MSE m5'] = data['MSE']
data['RMSE m5'] = data['RMSE']
data['MAPE m5'] = data['MAPE']

# # Sorting the dataframe by 'Ticket' and then by 'Model'
data= data[['Ticket', 'Model', 'MSE m5','RMSE m5', 'MAPE m5' ]]
sorted_data1 = data.sort_values(by=['Ticket', 'Model'])
# # Display the sorted dataframe
sorted_data1.head()
```
