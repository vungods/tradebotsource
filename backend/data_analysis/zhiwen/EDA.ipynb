```python
def plot_forex_data(data, currency_label):
    """
    Plot the forex data for a given currency pair with fixed titles.

    :param data: DataFrame containing the forex data
    :param currency_label: Label for the currency pair (e.g., "EUR/USD")
    """
    import matplotlib.pyplot as plt

    # Setting up the figure for Price Data plots (2x2 matrix)
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

    # Adjusted Close Price
    axes[0, 0].plot(data['datadate'], data['adjcp'], label='Adjusted Close Price', color='blue')
    axes[0, 0].set_title(f'{currency_label} Adjusted Close Price')
    axes[0, 0].set_xlabel('Date')
    axes[0, 0].set_ylabel('Price')
    axes[0, 0].legend()
    axes[0, 0].tick_params(axis='x', rotation=45)

    # Open Price
    axes[0, 1].plot(data['datadate'], data['open'], label='Open Price', color='green')
    axes[0, 1].set_title(f'{currency_label} Open Price')
    axes[0, 1].set_xlabel('Date')
    axes[0, 1].set_ylabel('Price')
    axes[0, 1].legend()
    axes[0, 1].tick_params(axis='x', rotation=45)

    # High Price
    axes[1, 0].plot(data['datadate'], data['high'], label='High Price', color='red')
    axes[1, 0].set_title(f'{currency_label} High Price')
    axes[1, 0].set_xlabel('Date')
    axes[1, 0].set_ylabel('Price')
    axes[1, 0].legend()
    axes[1, 0].tick_params(axis='x', rotation=45)

    # Low Price
    axes[1, 1].plot(data['datadate'], data['low'], label='Low Price', color='purple')
    axes[1, 1].set_title(f'{currency_label} Low Price')
    axes[1, 1].set_xlabel('Date')
    axes[1, 1].set_ylabel('Price')
    axes[1, 1].legend()
    axes[1, 1].tick_params(axis='x', rotation=45)

    # Adjusting layout for Price Data
    plt.tight_layout(pad=3.0)

    # Show the Price Data plots
    plt.show()

    # Creating a separate plot for Volume Data with rotated X-axis labels
    plt.figure(figsize=(12, 6))
    plt.bar(data['datadate'], data['volume'], color='gray')
    plt.title(f'{currency_label} Trading Volume')
    plt.xlabel('Date')
    plt.ylabel('Volume')
    plt.xticks(rotation=45)
    plt.show()

# Example usage of the function with the fixed titles
```

```python

# Example usage of the function
plot_forex_data(data_eu_m5, "EUR/USD 5-minute")
plot_forex_data(data_eu_h1, "EUR/USD 1-hour")
```

```python
plot_forex_data(data_eu_d1, "EUR/USD 1-day")
```
